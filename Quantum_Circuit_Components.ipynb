> ### Quantum Circuits

A quantum circuit is a model of quantum computation, that includes quantum registers, classical registers, quantum gates, barriers and measurements.

There are 5 essential components of a circuit:
<font color=orange>
- Quantum Registers

- Classical Registers

- Quantum Gates

- Barriers

- Measurements

</font>


> ### Quantum Registers
A quantum register is a collection of qubits on which gates and other operations act. Quantum computers perform calculations by manipulating qubits within a quantum register.
[QuantumRegister()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumRegister.html) is the function used for it. Click on it for further information on syntax and parameters.

> ### Classical Registers
A classical register consists of bits that can be written to and read within the coherence time of the quantum circuit.
Usually a qubit is measured and it's measured value is stored in a classical bit. If you do not understand the term "measurement" right now, don't worry we will cover this in the next part of this notebook.
[ClassicalRegister()](https://qiskit.org/documentation/stubs/qiskit.circuit.ClassicalRegister.html) is the function used for it. Click on it for further information on syntax and parameters.

<b><i>With this knowledge we can proceed to teach you to implement a quantum circuit on qiskit and initialize it. So let's begin!</i></b>

```python
#The following statements import useful libraries containing functions and methods that will help us build the circuit.
from qiskit import *
from qiskit_aer import Aer
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
import numpy as np
import matplotlib.pyplot as plt
print("Libraries imported successfully")
```

```python
qr=QuantumRegister(2)
cr=ClassicalRegister(2)
qc=QuantumCircuit(qr,cr)
q0,q1=qr
c0,c1=cr
qc.draw(output="mpl")
```


---

- The first line initializes a quantum register with 2 qubits using the <font color=darksalmon>QuantumRegister()</font> function.
- The second line initializes a classical register with 2 bits using the <font color=darksalmon>ClassicalRegister()</font> function.
- The third line initializes a quantum circuit with 2 qubits and 2 bits using the <font color=darksalmon>QuantumCircuit()</font> function.
- The <font color=darksalmon>draw()</font> function helps visualize the circuit.

---

Now we will apply gates, barriers and measurements to the circuit.


> ### Quantum Gates

A quantum gate is a basic quantum circuit operating on a small number of qubits. They are the building blocks of quantum circuits, like classical logic gates are for conventional digital circuits. Some examples of quantum gates are Pauli X gate using [circuit_name.x()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.x.html#qiskit.circuit.QuantumCircuit.x) , Hadamard gate using [circuit_name.h()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.h.html#qiskit.circuit.QuantumCircuit.h), and CNOT gate using [circuit_name.cx()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.cx.html#qiskit.circuit.QuantumCircuit.cx).

[Here is a list of all Quantum Logic Gates.](https://en.wikipedia.org/wiki/Quantum_logic_gate) We will dive deeper into each individual gate in the next tutorial

> ### Barriers
The barrier acts as a directive for circuit compilation to separate pieces of a circuit. [Barrier](https://qiskit.org/documentation/stubs/qiskit.circuit.library.Barrier.html) click here for more information.

> ### Measurements
In quantum physics, a measurement is the testing or manipulation of a physical system to yield a numerical result. [Measurement or measure()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.measure.html) can be used to yield a single classical numerical outcome and assign the value to a classical bit in the classical register.

> ### Reset
The reset function is used to set the qubits back to it's default state, i.e. 0. We use the [reset()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.reset.html)function.

<b><i>With this knowledge we can proceed to implement gates, barriers and measurements to our quantum circuit.</i></b>


```python
qc.x(q0)
qc.barrier()
qc.h(q1)
qc.measure(q0,c0)
qc.measure(q1,c1)
qc.draw(output="mpl")
```


---
- In the first line we apply a X gate using <font color=darksalmon>x()</font> on the first qubit.
- In the second line we use a barrier in our circuit using the <font color=darksalmon>barrier()</font> function.
- In the third line we apply a Hadamard gate using <font color=darksalmon>h()</font> on the second qubit.
- In the fourth and fifth lines we measure each qubit's outcome into it's respective classical bitusing the <font color=darksalmon>measure()</font> function.
---

```python
qr=QuantumRegister(2)
cr=ClassicalRegister(2)
qc=QuantumCircuit(qr,cr)
q0,q1=qr
c0,c1=cr
qc.x(q0)
qc.barrier()
qc.h(q1)
qc.reset(q0)
qc.measure(q0,c0)
qc.measure(q1,c1)
qc.draw(output="mpl")
```

> ### Histogram
A histogram is an approximate representation of the distribution of numerical data. Here it helps us visualize the outcome of our quantum circuit. We use the [plot_histogram()](https://qiskit.org/documentation/stubs/qiskit.visualization.plot_histogram.html#qiskit.visualization.plot_histogram) function to generate histograms on qiskit.


> ### Bloch Sphere
- The Bloch sphere is a unit sphere, where the north and south poles represent the basis states |0⟩ and |1⟩, respectively. The points on the surface of the sphere represent superposition states, which are combinations of |0⟩ and |1⟩.

- By mapping the amplitudes of a qubit's state onto the Bloch sphere, we can easily visualize and analyze the qubit's behavior, such as rotations and transformations, as well as gain insights into quantum operations and quantum gates.

- We use the [plot_bloch_vector()](https://qiskit.org/documentation/stubs/qiskit.visualization.plot_bloch_vector.html) to generate bloch spheres. You can learn more on Bloch Sphere here "add link to bloch sphere tutorial "




```python
#Measurement of all qubits
qc_measured = qc.measure_all(inplace=False)

#Measurements results
from qiskit.primitives import StatevectorSampler
sampler = StatevectorSampler()
job = sampler.run([qc_measured], shots=1000)
result = job.result()
print("\nMeasurement results:")
print(f"Counts: {result[0].data['meas'].get_counts()}")

print("\nBar Chart:")
plot_histogram(result[0].data.meas.get_counts())
```


---
- The above histogram has been generated using <font color=darksalmon>Basic Aer simulator.</font> This gives us the outcome of our circuit and informs us what state each qubit is in.

- Notice that the 1 on top of the outcomes |01⟩ and |11⟩ is the first qubits result. And the 0 and 1 are the results of the second qubit. Remember that qiskit uses <font color=darksalmon>little endian ordering</font> [hyperlink to endian ordering blog]. In qiskit the default state of a qubit is zero. Since we applied the x gate on the first qubit it changes the qubit's state to 1. Since we applied the hadamard gate on second qubit, it's in superposition [insert superposition blog hyperlink] and can be both 0 and 1. Hence one outcome with it being 0 and other outcome with it being 1.
