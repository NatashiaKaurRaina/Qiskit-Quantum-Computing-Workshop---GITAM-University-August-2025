> ### Quantum Gates
- In quantum computing and specifically the quantum circuit model of computation, a quantum logic gate (or simply quantum gate) is a basic quantum circuit operating on a small number of qubits.

- They are the building blocks of quantum circuits, like classical logic gates are for conventional digital circuits.Unlike many classical logic gates, quantum logic gates are reversible.

#### <b><i>X Gate</b></i>
The <font color=darksalmon>X gate</font> behaves like a classical NOT gate. It flips the state of a qubit from 0 to 1 or vice versa. It performs a 180 degree rotation about the x-axis. Using the [x()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.x.html#qiskit.circuit.QuantumCircuit.x) function. Click on it for more information.



<center>
\(\lvert 0 \rangle \xrightarrow{\,X\,} \lvert 1 \rangle\)    ;      
  \(\lvert 1 \rangle \xrightarrow{\,X\,} \lvert 0 \rangle\)

$\ket{0}-X->\ket{1}$

$\ket{1}-X->\ket{0}$

</center>

```python
plot_bloch_multivector(qc)
```

#### <b><i>Y Gate</b></i>
The <font color=darksalmon>Y gate</font> performs a 180 degree rotation about the y-axis. It converts 0 to 1 and vice versa just like the X gate but also performs phase flip as shown below. The Y Gate is also known as Bit and Phase flip Gate. This is because it performs a bit flip like X Gate and a Phase flip like Z Gate. Click on [y()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.y.html#qiskit.circuit.QuantumCircuit.y) for more information.


<center>
\(\lvert 0 \rangle \xrightarrow{\,Y\,} i\lvert 1 \rangle\)    ;      
  \(\lvert 1 \rangle \xrightarrow{\,X\,} -i\lvert 0 \rangle\)

$\ket{0}-Y-> i \ket{1}$

$\ket{1}-Y-> -i \ket{0}$

</center>

---

#### <b><i>Z Gate</b></i>
The <font color=darksalmon>Z gate</font> performs a 180 degree rotation about the z-axis. The gate is equivalent to a phase flip. Click on [z()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.z.html#qiskit.circuit.QuantumCircuit.z) for more information.


<center>
\(\lvert 0 \rangle \xrightarrow{\,Z\,} \lvert 0 \rangle\)    ;      
  \(\lvert 1 \rangle \xrightarrow{\,Z\,} -\lvert 1 \rangle\)

$\ket{0}-Z->\ket{0}$

$\ket{1}-Z-> - \ket{1}$

</center>

---

#### <b><i>Hadamard Gate</b></i>
The <font color=darksalmon>H gate</font> puts a qubit into superposition states $\ket{+} or \ket{-}$. Click on [h()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.h.html#qiskit.circuit.QuantumCircuit.h) for more information.


<center>
\(\lvert 0 \rangle \xrightarrow{\,H\,} \lvert + \rangle\)    ;      
  \(\lvert 1 \rangle \xrightarrow{\,H\,} \lvert - \rangle\)

$\ket{0}-Z->\ket{+}$

$\ket{1}-Z->\ket{-}$

</center>

#### <b><i>Identity Gate</b></i>

The <font color=darksalmon>global phase gate</font> introduces a global phase $e^{i\varphi}$ to the whole qubit quantum state. A quantum state is uniquely defined up to a phase.

### CX Gate
- The <font color=darksalmon>CNOT gate</font> is two-qubit gate, where the first qubit is the control qubit and the second qubit is the target qubit. If the control qubit is 0 then the target qubit doesn't change. If the control qubit is 1 then the target qubit state is flipped from 0 to 1 or vice versa.

- Hence performing the X gate operation on the target qubit if the control qubit is 1. Leading to the gate getting it's name <font color=darksalmon>Controlled X gate.</font> Using the [cx()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.cx.html#qiskit.circuit.QuantumCircuit.cx) function.

- Using little endian odering, the right qubit being the first qubit and the left qubit being the second qubit. $\ket{qubit_2,qubit_1}$ let's see what the CX or CNOT gate does.

<center>

$\ket{0,0}-CX->\ket{0,0}$

$\ket{1,0}-CX->\ket{1,0}$

$\ket{0,1}-CX->\ket{1,1}$

$\ket{1,1}-CX->\ket{0,1}$

</center>

---

### CZ Gate
The <font color=darksalmon>CZ gate or CPhase gate</font> is two-qubit operation where the first qubit is the control qubit and the second qubit is the target qubit. It performs the Z gate operation on the target qubit if the control qubit is 1. Leading to the gate getting it's name <font color=darksalmon>Controlled Z gate.</font> Using the [cz()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.cz.html#qiskit.circuit.QuantumCircuit.cz) function.

<center>

$\ket{0,0}-CZ->\ket{0,0}$

$\ket{1,0}-CZ->\ket{1,0}$

$\ket{0,1}-CZ->\ket{0,1}$

$\ket{1,1}-CZ->\ket{-1,1}$

</center>

---

### Rxx Gate

The <font color=darksalmon>RXX gate</font> implements $e^{-i\theta/2 X \otimes X}$. The MlmerSrensen gate, the native gate on ion-trap systems, can be expressed as a sum of RXX gates. Using the [rxx()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.rxx.html#qiskit.circuit.QuantumCircuit.rxx) function.

---

### Ryy Gate
The <font color=darksalmon>RYY gate</font> implements $e^{-i\theta/2 Y \otimes Y}$. Using the [ryy()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.ryy.html#qiskit.circuit.QuantumCircuit.ryy) function.

---

### Rzz Gate
The <font color=darksalmon>RZZ gate</font> implements $e^{-i\theta/2 Z \otimes Z}$. The RZZ gate requires a single parameter: an angle expressed in radians. This gate is symmetric; swapping the two qubits it acts on doesnt change anything. Using the [rzz()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.rzz.html#qiskit.circuit.QuantumCircuit.rzz) function.

---

### Rxy Gate
The <font color=darksalmon>RXY gate</font> implements $e^{-i\theta/4 (X \otimes X + Y \otimes Y)}$. Using the [rxy()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.rxy.html#qiskit.circuit.QuantumCircuit.rxy) function.

---


```python
qr=QuantumRegister(2)
cr=ClassicalRegister(2)
qc=QuantumCircuit(qr,cr)
q0,q1=qr
c0,c1=cr
qc.x(q1)
qc.cx(q0,q1)
qc.draw(output="mpl")
```

```python
plot_bloch_multivector(qc)
```

```python
qr=QuantumRegister(2)
cr=ClassicalRegister(2)
qc=QuantumCircuit(qr,cr)
q0,q1=qr
c0,c1=cr
qc.rxx(pi/4,q0,q1)
qc.draw(output="mpl")
```

```python
plot_bloch_multivector(qc)
```

### Toffoli Gate or CCX Gate
<font color=darksalmon>Toffoli gate</font>, also known as the CCX gate flips the state of the target qubits if the two control qubits are in the $\ket{11}$ state. Just like the CX gate but with two control qubits. Using the [ccx()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.ccx.html#qiskit.circuit.QuantumCircuit.ccx) function.

<center>

$\ket{0,0,0}-CCX->\ket{0,0,0}$

$\ket{1,0,0}-CCX->\ket{1,0,0}$

$\ket{0,1,0}-CCX->\ket{0,1,0}$

$\ket{0,0,1}-CCX->\ket{0,0,1}$

$\ket{0,1,1}-CCX->\ket{1,1,1}$

$\ket{1,0,1}-CCX->\ket{1,0,1}$

$\ket{1,1,0}-CCX->\ket{1,1,0}$

$\ket{1,1,1}-CCX->\ket{0,1,1}$

</center>

---

### Swap Gate
The <font color=darksalmon>SWAP gate</font> is two-qubit operation. The SWAP gate swaps the state of the two qubits involved in the operation. Using the [swap()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.swap.html#qiskit.circuit.QuantumCircuit.swap) function.

---

### iSwap Gate
The <font color=darksalmon>iSwap gate's</font> action is to swap two qubit states and phase the $\ket{01}$ and $\ket{10}$ amplitudes by i. Using the [i()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.i.html#qiskit.circuit.QuantumCircuit.i) function.

---

### CSwap Gate
<font color=darksalmon>Controlled-SWAP gate</font>, also known as the Fredkin gate swaps the states of the two target qubits if the control qubit is in the $\ket{1}$ state. Using the [cswap()](https://qiskit.org/documentation/stubs/qiskit.circuit.QuantumCircuit.cswap.html#qiskit.circuit.QuantumCircuit.cswap) function.

----


```python
qr=QuantumRegister(3)
cr=ClassicalRegister(3)
qc=QuantumCircuit(qr,cr)
q0,q1,q2=qr
c0,c1,c2=cr
qc.x(q0)
qc.ccx(q0,q1,q2)
qc.draw(output="mpl")
```

```python
plot_bloch_multivector(qc)
```

```python
qr=QuantumRegister(2)
cr=ClassicalRegister(2)
qc=QuantumCircuit(qr,cr)
q0,q1=qr
c0,c1=cr
qc.x(q0)
qc.swap(q0,q1)
qc.draw(output="mpl")
```

```python
plot_bloch_multivector(qc)
```
